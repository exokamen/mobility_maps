## Amsterdam

```python
col = '#FF0033'

m = folium.Map(location=[52.3676, 4.9041], tiles='cartodbpositron', zoom_start=12.5)
# folium.TileLayer('cartodbdark_matter').add_to(m)
# folium.TileLayer('stamenterrain').add_to(m)
# folium.TileLayer('stamentoner').add_to(m)


parks = r'C:\Users\TodorovK\OneDrive - Rebelgroup\Desktop\fun_maps\amsterdam_parks.geojson'


parking_spaces_2019 = 156593
area_sq_m = parking_spaces_2019 * 15
  
    
choropleth = folium.Choropleth(geo_data=parks, 
                               fill_color='Green', 
                               opacity= 0.35, 
                               weight=0.5, 
                               key_on='feature.properties.Naam'
                              ).add_to(m)

folium.LayerControl().add_to(m)
choropleth.geojson.add_child(
    folium.features.GeoJsonTooltip(['Naam'], labels=False)
)

folium.Circle(
       location=[52.3676, 4.9041],
       tooltip='{} parking spaces (2019)'.format(parking_spaces_2019),
       radius=np.sqrt(area_sq_m/np.pi),
       color=col,
       fill=True,
       opacity= 0.79,
       fill_color=col,
       stroke= True,
       weight= 1
     ).add_to(m)

# folium.TopoJson(
#     open(parks),
#     name='Parks',
# ).add_to(m)


# m.add_child(folium.GeoJson(data = open(parks)))

# folium.GeoJson(
#     parks,
#     name='Parks'
#     ).add_to(m)

# folium.Choropleth(geo_data=nbrhoods, data=population2016,
#              columns=['nbhd_name', 'Population, 2016'],
#              key_on='feature.properties.AREA_NAME',
#              threshold_scale=[0, 50000, 100000, 150000, 200000],
#              fill_color='BuPu', 
#              legend_name='Population (2016)').add_to(m)


folium.LayerControl().add_to(m)       
m.save('amsterdam_parking.html')
m
```

```python
area_sq_km = area_sq_m / 1e6
print('Parking area {:.2f}km^2'.format(area_sq_km))

vondelpark_area_sq_km = 0.485623
print('{:.1f} times larger than Vondelpark!'.format(area_sq_km / vondelpark_area_sq_km))

print('{:.0f} apartments of 60m^2!'.format(area_sq_m / 60.0))

print('{:.1f}bn Euro at 5 000 Euro per m^2'.format(area_sq_m * 5000 / 1e9))
```

## Den Haag

```python
col = '#FF0033'

m = folium.Map(location=[52.3676, 4.9041], tiles='cartodbpositron', zoom_start=13.5)
folium.TileLayer('cartodbdark_matter').add_to(m)
folium.TileLayer('stamenterrain').add_to(m)
folium.TileLayer('stamentoner').add_to(m)
folium.LayerControl().add_to(m)

parking_spaces_2019 = 156593
area_sq_m = parking_spaces_2019 * 15
        
folium.Circle(
       location=[52.3676, 4.9041],
       popup='Parking spots combined area',
       radius=np.sqrt(area_sq_m/np.pi),
       color=col,
       fill=True,
       opacity= 0.79,
       fill_color=col,
       stroke= True,
       weight= 1
     ).add_to(m)
        
m.save('amsterdam_parking.html')

m
```
